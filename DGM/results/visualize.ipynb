```python
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib import rcParams
from analyze import load_dimension_data, load_activation_data

%matplotlib inline
rcParams.update({"legend.fontsize": 14, 
            "legend.loc": "upper right", 
            "figure.titlesize": "xx-large",
            "figure.figsize": (20, 18),
            "axes.titlesize": "x-large",
            "axes.grid": True,
            "grid.linestyle": ":"})
```

```python
# Normalized Losses

loss2D = load_dimension_data(2, "loss")
loss4D = load_dimension_data(4, "loss")
loss8D = load_dimension_data(8, "loss")
loss12D = load_dimension_data(12, "loss")

lossTanh = load_activation_data("tanh", "loss")
lossSig = load_activation_data("sigmoid", "loss")
lossLogSig = load_activation_data("logsigmoid", "loss")
lossHardSig = load_activation_data("hardsigmoid", "loss")
lossSoft = load_activation_data("softplus", "loss")
```

### Base Parameters

* Epochs: 1250
* Time-Space Batch: 1000
* Boundary Batch: 100
* Boundary Penalty: 100
* Differential Operator Step: .0001
* ADAM Learning Rate: .0001
* ADAM Weight Decay: .01 


```python
fig, axs = plt.subplots(2, 2)

fig.suptitle("Loss per Epoch by Dimension")
loss2D.plot(ax = axs[0,0], title = "2-Dimensions", legend = False)
loss4D.plot(ax = axs[0,1], title = "4-Dimensions", legend = False)
loss8D.plot(ax = axs[1,0], title = "8-Dimensions", legend = False)
loss12D.plot(ax = axs[1,1], title = "12-Dimensions", legend = False)

lines, labels = fig.axes[0].get_legend_handles_labels()
fig.legend(lines, labels)

fig.text(0.5, 0.1, "Epoch", ha = "center", fontsize = 14)
fig.text(0.1, 0.5, "Loss", ha = "center", rotation = "vertical", fontsize = 14)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

fig.suptitle("Loss per Epoch by Activation")
lossTanh.plot(ax = axs[0,0], title = "Tanh", legend = False)
lossSig.plot(ax = axs[0,1], title = "Sigmoid", legend = False)
lossHardSig.plot(ax = axs[1,0], title = "Hard Sigmoid", legend = False)
lossSoft.plot(ax = axs[1, 1], title = "Softplus", legend = False)


lines, labels = fig.axes[0].get_legend_handles_labels()
fig.legend(lines, labels)

fig.text(0.5, 0.1, "Epoch", ha = "center", fontsize = 14)
fig.text(0.1, 0.5, "Loss", ha = "center", rotation = "vertical", fontsize = 14)
plt.show()
```
